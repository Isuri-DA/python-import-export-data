# 2. Read CSV file using csv.DictReader

```python
import csv

with open("./data/Artificial Movies Dataset.csv", "r") as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        print(row)
        
# Output -  maps the information in each row to a dict
# {'Title': 'The Silence of the Lambs', 'Rating': '8.2', 'BoxOfficeCollection': '418357897', 
# 'Budget': '155958062', 'Director': 'Steven Spielberg', 'Producer': 'George Lucas'}
```
